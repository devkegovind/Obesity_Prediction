```python
def bar_plot(df):
    for col in df[cat_columns]:
        plt.bar(df[col], df['Age'])
        plt.xlabel(col)
        plt.ylabel('Age')
        plt.title(f'{col} vs Age Bar Plot')
        plt.show()
```

```python
bar_plot(df)
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
cat_columns
```

```python
def wordcloud(df, cat_columns):
    for col in cat_columns:
        text = " ".join(str(i) for i in df[col])
        stopwords = set(STOPWORDS)
        wordcloud = WordCloud(stopwords = stopwords, background_color="black").generate(text)
        plt.style.use('classic')
        plt.figure(figsize=(12,10))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.tight_layout()
        plt.show()
```

```python
wordcloud(df, cat_columns)
```

```python
def pie_chart(df, cat_columns):
    for col in cat_columns:
        plt.pie(df[col].value_counts(),labels=df[col].unique(), autopct = '%1.1f%%')
        plt.title(f"Pie Chart of {col}")
        plt.show()
```

```python
pie_chart(df, cat_columns)
```

```python
def heatmap(df, cat_columns):
    for col in cat_columns:
        sns.heatmap(data=pd.crosstab(index=df[col], columns=df[col]))
        plt.show()
```

```python
heatmap(df, cat_columns)
```

```python
def hist_plot(df, num_columns):
    for col in num_columns:
        plt.hist(df[col], bins = 20, edgecolor='k')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.title(f"Histogram of {col}")
        plt.show()
```

```python
hist_plot(df, num_columns)
```

```python
sns.pairplot(df[num_columns])
plt.title("Pair Plot")
plt.show()
```

```python
correlation_matrix = df[num_columns].corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python

def kde_plot(df, num_columns):
    for col in num_columns:
        sns.kdeplot(df[col], shade = True)
        plt.xlabel(col)
        plt.ylabel('Density')
        plt.title(f"Density Plot of {col}")
        plt.show()
```

```python
kde_plot(df, num_columns)
```
