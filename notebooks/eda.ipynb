```python
cat_columns = df.columns[df.dtypes == 'object']
cat_columns
```

```python
def bar_plot(df):
    for col in df[cat_columns]:
        plt.bar(df[col], df['Age'])
        plt.xlabel(col)
        plt.ylabel('Age')
        plt.title(f'{col} vs Age Bar Plot')
        plt.show()
```

```python
bar_plot(df)
```
