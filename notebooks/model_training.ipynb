```python
X = df.drop('NObeyesdad', axis = 1)
y = df['NObeyesdad']
```

```python
from sklearn.model_selection import train_test_split

# Split Dataset into Training & Test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Standard Scale Data
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

# Tested MinMaxscaler as KNN Historicall Does Better with MinMax
mm = MinMaxScaler()
X_train_scaled_mm = mm.fit_transform(X_train_scaled)
X_test_scaled_mm = mm.transform(X_test_scaled)

def model_predict(X_train: pd.DataFrame, y_train: pd.DataFrame,X_test: pd.DataFrame, y_test: pd.DataFrame):
    models = [
        ('Random Forest', RandomForestClassifier()),
        ('Decision Tree', DecisionTreeClassifier()),
        ('KNN', KNeighborsClassifier()),
        ('SVM', SVC())
    ]

    for name, model in models:
        # Unscaled Data
        clf = model.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        # Scaled Data
        clf_scaled = model.fit(X_train_scaled, y_train)
        y_pred_scaled = clf_scaled.predict(X_test_scaled)

        # mm scaled Data
        clf_scaled_mm = model.fit(X_train_scaled_mm, y_train)
        y_pred_scaled_mm = clf_scaled_mm.predict(X_test_scaled_mm)

        # Accuracy Scores
        accuracy = round(accuracy_score(y_test, y_pred),5)
        scaled_acc = round(accuracy_score(y_test, y_pred_scaled),5)
        scaled_mm_acc = round(accuracy_score(y_test, y_pred_scaled_mm),5)

        # Output
        print(name + ':')
        print("************************************")
        print(f"Accuracy : {accuracy}")
        print(f"Accuracy w/Scaled Data(ss):{scaled_acc}")
        print(f"Accuracy w/Scaled Data(mm):{scaled_mm_acc}")
        if (accuracy > scaled_acc) and (accuracy > scaled_mm_acc):
            print(f"\nClassification Report:\n{classification_report(y_test, y_pred)}")
            print('***********************************************')
        elif (scaled_acc > scaled_mm_acc):
            print(f"\nClassification Report:\n{classification_report(y_test, y_pred_scaled_mm)}")
            print("********************************************")
        else:
            print(f"\nClassification Report\n:{classification_report(y_test, y_pred_scaled_mm)}")
            print("********************************************")
```

```python
model_predict(X_train, y_train, X_test, y_test)
```
